## Analyzing and Picking top 5 Stocks

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
 
```

```python
# read csv file 
stocks = pd.read_csv('../data/stocks_closing_info_2015.csv').set_index('Ticker Symbol').round(2)
stocks
```

```python
# Filter list to stock price is smaller than $100
priced_filtered = stocks.loc[
    stocks['Closing Price'] < 100,
    :
]
# priced_filtered["Closing Price"] = priced_filtered["Closing Price"].map("% {:,.2f}".format)
```

```python
priced_filtered
```

## Working Capital Ratio 

- Deciding which company stocks are more liquide 

- Ratio should be between 1.5- 2

```python
# working capital ratio = current assets / current liabilities
priced_filtered['Working Capital Ratio'] = priced_filtered['Total Current Assets'] / priced_filtered['Total Current Liabilities']
```

```python
working_capital = pd.DataFrame(priced_filtered['Working Capital Ratio']).sort_values(by='Working Capital Ratio', ascending=False).round(2)

working_capital_top = working_capital.loc[
    priced_filtered['Working Capital Ratio'] < 2,
    :
]
working_capital_top.head()
```

## Quick Ratio 

- Considered to be more consertvative than current ratio

- The higher the ratio result, the better a company's liquidity

```python
quick_ratio = pd.DataFrame(priced_filtered['Quick Ratio']).sort_values(by='Quick Ratio', ascending=False).round(2)
quick_ratio.head()
```

## Earnings Per Share (EPS)

- A higher EPS indicates greater value because investors will pay more for a company's shares if they think the company has higher profits relative to its share price.

```python
earnings_per_share = pd.DataFrame(priced_filtered['Earnings Per Share']).sort_values(by='Earnings Per Share', ascending=False).round(2)
earnings_per_share.head()
```

## Price to Earnings Ratio (P/E)

- Investors expect higher earnings with higher P/E 

```python
# Earnings Per Share/ closing price = percentage - top 5 stocks 
priced_filtered['P/E Ratio'] = priced_filtered['Closing Price']/priced_filtered['Earnings Per Share'] 
earnings_vs_closing = pd.DataFrame(priced_filtered['P/E Ratio']).sort_values(by='P/E Ratio', ascending=False).head()
earnings_vs_closing['P/E Ratio'] = earnings_vs_closing['P/E Ratio'].map(" {:,.2f}".format)
earnings_vs_closing
```

## Debt to Equity ratio 

```python
# Find shareholders equity
priced_filtered['Shareholders Equity'] = priced_filtered['Total Assets']-priced_filtered['Total Liabilities']
priced_filtered
```

```python
# Debt to equity is total liabilities / Shareholder Equity
priced_filtered['D/E Ratio'] = priced_filtered['Total Liabilities'] / priced_filtered['Shareholders Equity']
```

```python
# D/E should not be above 2 - 5 top slowest stocks 
debt_vs_equity = pd.DataFrame(priced_filtered['D/E Ratio']).sort_values(by='D/E Ratio', ascending=True).round(2).head()
debt_vs_equity
```
